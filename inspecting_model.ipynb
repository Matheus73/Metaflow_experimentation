# Inspecting Flows and Results


This notebook is used to receive te datas from the last execution of modelTest.py (predictions and measures), print those values and make graphs

## import the metaflow client

```python
from metaflow import Flow,get_metadata,Step,Run
import matplotlib.pyplot as plt
import seaborn as sns
print("Current metadata provider: %s" % get_metadata())
```

## Get measures from the lastest run of ModelTest

```python
run = Flow('ModelTest').latest_successful_run
           
print("Using run: %s" % str(run))
```

```python
measures = run.data.measures
measures
```

## Graph of predictions

```python
answers = run.data.compare['answers']
results = run.data.compare['results']
plt.scatter(answers,results)
```

```python
sns.distplot(answers-results,bins=50)
```
